```python
import hashlib as hasher
import datetime as date
import time
```

```python
#define block in a blockchain

class Block:
    def __init__(self, index, timestamp, data, previous_hash=''):
        self.nonce = 0
        self.index = index
        self.timestamp = timestamp
        self.data = data
        self.previous_hash = previous_hash
        self.hash = self.hash_block()
        
    def hash_block(self):
        a_string = (str(self.nonce) + str(self.index) + str(self.timestamp) + str(self.data) + str(self.previous_hash))
        hashed_string = hasher.sha256(a_string.encode('utf-8')).hexdigest()
        return hashed_string
    
    def mineBlock(self,difficulty):
        while(self.hash[:difficulty] != '0000'):
            self.nonce+=1
            self.hash = self.hash_block()
        
        print ("Timestamp: {}".format(self.timestamp))
        print ("Data: {}".format(self.data))
        print ("Previous hash: {}".format(self.previous_hash))
        print ("Current hash calculated: {}".format(self.hash))
        print ("Corresponding nonce: {}\n".format(self.nonce))
    

class Blockchain:
    def __init__(self):
        self.chain = []
        self.chain.append(self.createGenesisBlock())

    
    def createGenesisBlock(self):
        return Block(0, date.datetime.now(), "First block", "0")
    
    def getlatestBlock(self):
        return self.chain[len(self.chain) - 1]
    
#     def next_block(last_block):
#         new_index = last_block.index + 1
#         new_timestamp = date.datetime.now()
#         new_data = "hey this is the new block" + str(new_index)
#         new_hash = last_block.hash
#         current_hash = last_block.mineBlock(5)
    
#         return Block(new_index, new_timestamp, new_data)

    
    def addBlock(self,newBlock):
        newBlock.previous_hash = self.getlatestBlock().hash
        newBlock.mineBlock(4)
        self.chain.append(newBlock)
        
    def ischainValid(self):
        for i in range(len(self.chain)):
            currentBlock = self.chain[i]
            previousBlock = self.chain[i-1]
            
            if (currentBlock.previous_hash != previousBlock.hash):
                print ("block# " + str(i-1) + "& " + "block# " + str(i) + " are not valid")
            else:
                print ("block# " + str(i-1) + "& " + "block# " + str(i) + " are valid")


# def create_genesis_block():
#     return Block(0, date.datetime.now(), "First block", "0")
    
# def next_block(last_block):
#     new_index = last_block.index + 1
#     new_timestamp = date.datetime.now()
#     new_data = "hey this is the new block" + str(new_index)
#     new_hash = last_block.hash
#     current_hash = last_block.mineBlock(2)
    
#     return Block(new_index, new_timestamp, new_data)

    
# # def addBlock(self,newBlock):
# #     new_index = last_block.index + 1
# #     new_timestamp = date.datetime.now()
# #     new_data = "hey this is the new block" + str(new_index)
# #     newBlock.previous_hash = last_block.hash
# #     newBlock.mineBlock(2)
# #     self.chain.append(newBlock)
        
# def ischainValid(self):
#     for i in range(len(chain)):
#         currentBlock = self.chain[i]
#         previousBlock = self.chain[i-1]
            
#         if(currentBlock.hash != currentBlock.hash_block()):
#             return false
        
#         if(currentBlock.previous_hash != previousBlock.hash):
#             return false
        
#         return true
```
